```python
y=[1,4,9,16,25,36,49,64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
l1=ax.plot(x1,y,'gs-')
l2=ax.plot(x2,y,'ro-')
ax.legend(labels = ('Java', 'Python'), loc = 'lower right')
ax.set_title('Best Programming Language')
ax.set_xlabel('Software Development')
ax.set_ylabel('Machine Learning')
```
